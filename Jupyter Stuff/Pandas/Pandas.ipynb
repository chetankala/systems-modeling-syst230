```python
data = {"Name": ["Jack", "Ahmad", "John", "Elise", "Xu"],
       "Age": [20, 24, 32, 42, 51],
       "Salary": [100000, 93000, 88000, 120000, 94000],
       "Job": ["Designer", "Chef", "Programmer", "Secretary", "IT"]}

df = pd.DataFrame(data)
display(df)
```

```python
filtered_values = df.loc[(df["Salary"]>93000) & (df["Age"]>40), ["Salary", "Age"]]
display(filtered_values)
```

```python
# numpy.where()

filtered_values = np.where((df["Salary"]>93000) & (df["Age"]>40))
display(df.loc[filtered_values])
```

```python
filtered_values = df.query("Salary > 93000 & Age > 40")
display(filtered_values)
```

```python
filtered_values = df[df.eval("Salary > 93000 & Age > 40")]
display(filtered_values)
```
