```python
import numpy as np

x = np.random.randn(10000)
plt.hist(x, bins = 100, color = "red", histtype = "stepfilled", cumulative = -1, label = "Randome Values")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Histogram")
plt.show()
```

```python
Data1 = np.random.normal(-5, 5, 10000)
Data2 = np.random.normal(-10, 10, 10000)
plt.hist(Data1, bins = 100, color = "green", alpha = 0.8, label = "Data1")
plt.hist(Data2, bins = 100, color = "purple", alpha = 0.4, label = "Data2")
plt.legend()
plt.title("Histogram")
plt.show()
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
markersize = 1000 * np.random.randn(100)

plt.scatter(x, y, c = colors, s = markersize, cmap = "cividis", alpha = 0.3)
plt.colorbar()
plt.show()
```

```python
x_1 = np.random.randn(100)
y_1 = np.random.randn(100)

x_2 = np.random.randn(100)
y_2 = np.random.randn(100)

plt.scatter(x_1, y_1, color = "yellow", marker = "s", linewidth = 3,
           edgecolor = "green", label = "Data1")

plt.scatter(x_2, y_2, color = "blue", marker = "H", linewidth = 3,
           edgecolor = "red", label = "Data2")

plt.xlabel("Time")
plt.ylabel("Data")
plt.legend()
plt.title("Scatter Plot")
plt.show()
```

```python
# subplots(nrows, ncols)
# fig, ax = plt.subplots(2, 2)

x = np.random.randn(10)
y = x*5

fig, ax = plt.subplots(2, 2)

ax[0][0].plot(x, y, color = "orange")
ax[0][0].set_title("Line")

ax[0][1].bar(x, y, color = "gray")
ax[0][1].set_title("Bar Plot")

ax[1][0].scatter(x, y, color = "black")
ax[1][0].set_title("Scatter Plot")

ax[1][1].hist(y, bins = 10)
ax[1][1].set_title("Histogram")

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from itertools import count
import random

x_vals = list()
y1_vals = []
y2_vals = []

index = count()

def gen_data(i):
    x_vals.append(next(index))
    y1_vals.append(random.randint(-5, 10))
    y2_vals.append(random.randint(-6, 12))
    
    plt.cla()
    
    plt.plot(x_vals, y1_vals, label = "Data 1")
    plt.plot(x_vals, y2_vals, label = "Data 2")
    plt.legend(loc = "upper left")
    
    plt.tight_layout()
    
anim = FuncAnimation(plt.gcf(), gen_data, interval = 1000)

plt.tight_layout()
plt.show()
```
